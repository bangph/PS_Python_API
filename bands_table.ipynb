## Bands and wavelengths Lookup Table

In this Notebook we create a FITS Table that contains the band name and the corresponding band wavelength in micrometers.

Developed by - Anik Halder

```python
import numpy as np

import re

# reading the metadata.txt file

filename = "metadata.txt"

f = open(filename, 'r')

lines = f.readlines()
```

```python
band_name = np.ones(0) # band name array
WV = np.ones(0) # wavelengths array

i = 471 # Band 1 index in file 
j = 0 # indexing for bands and wavelengths

# 2659 is the index for the wavelength of the last band i.e. 438
while i<=2659:
    
    # extracting band number
    l = lines[i].split()
    # adding the band to the bands array
    band_name = np.append(band_name, int(l[1])) 
    
    i += 3 # going to the line to extract wavelength
    
    # extracting wavelength of the corresponding band number
    l = lines[i].split()
    # adding the wavelength to the wavelengths array
    WV = np.append(WV, float(re.findall("\d+\.\d+", l[0])[0]))
    
    i += 2 # going to the next iteration to extract band number
    j += 1 # next iteration index to store in bands and wavelengths
```

```python
from astropy.io import fits

from astropy.table import Table, Column

# defining table that will hold information about band_names and wavelengths

b_name = Column(band_name, name='band_name', dtype='int') # column having band name
w_lengths = Column(WV, name='wavelength', dtype='float') # column having wavelengths

bands_table = Table((b_name,w_lengths))  

bands_table['band_name'].unit = ''
bands_table['wavelength'].unit = 'Micrometers'

bands_table.show_in_notebook()
```

We have thus created a FITS Table - "bands_table.fits" - that contains the 438 different bands extracted from the 'metadata.txt' file along with the corresponding wavelengths in micrometers. Here on, we shall use this FITS file to look up the band and wavelength data in the other programs.
